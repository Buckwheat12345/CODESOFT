```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split

# Sample Movie Ratings Dataset
data = {
    'User': ['User1', 'User1', 'User2', 'User2', 'User3', 'User3', 'User4', 'User4'],
    'Movie': ['Movie1', 'Movie2', 'Movie1', 'Movie2', 'Movie2', 'Movie3', 'Movie3', 'Movie4'],
    'Rating': [5, 4, 3, 5, 4, 2, 1, 3]
}

df = pd.DataFrame(data)

# Create a user-movie matrix
user_movie_matrix = df.pivot_table(index='User', columns='Movie', values='Rating', fill_value=0)

# Split the data into training and testing sets
train_data, test_data = train_test_split(user_movie_matrix, test_size=0.2, random_state=42)

# Calculate cosine similarity between users
user_similarity = cosine_similarity(train_data)

# Function to predict ratings using collaborative filtering
def predict_ratings(user_movie_matrix, user_similarity):
    predicted_ratings = pd.DataFrame(index=user_movie_matrix.index, columns=user_movie_matrix.columns, dtype='float64')

    for i in range(len(user_movie_matrix.index)):
        for j in range(len(user_movie_matrix.columns)):
            if user_movie_matrix.iloc[i, j] == 0:
                # Calculate the weighted sum of ratings from similar users
                numerator = sum(user_similarity[i, k] * user_movie_matrix.iloc[k, j] for k in range(len(user_movie_matrix.index)))
                # Calculate the sum of absolute similarities
                denominator = sum(abs(user_similarity[i, k]) for k in range(len(user_movie_matrix.index)))

                # Predict the rating
                predicted_ratings.iloc[i, j] = numerator / (denominator + 1e-8)

    return predicted_ratings

# Predict ratings
predicted_ratings = predict_ratings(train_data, user_similarity)

# Function to recommend movies for a given user
def recommend_movies(user, user_movie_matrix, predicted_ratings):
    # Get unrated movies for the given user
    unrated_movies = user_movie_matrix.columns[user_movie_matrix.loc[user] == 0]

    # Sort the unrated movies based on predicted ratings
    recommendations = predicted_ratings.loc[user, unrated_movies].sort_values(ascending=False)

    return recommendations

# Example usage
user_to_recommend = 'User1'
recommendations = recommend_movies(user_to_recommend, user_movie_matrix, predicted_ratings)

print(f"Recommended movies for {user_to_recommend}:")
print(recommendations)
```
